### Test score visualization by gender

```python
fig, axes = plt.subplots(1, 3)
sns.boxplot(x="gender",y="math score" ,data=data, palette="colorblind",ax=axes[0]).set_title('Math Score by Gender')
sns.boxplot(x="gender",y="reading score" ,data=data, palette="colorblind",ax=axes[1]).set_title('Reading Score by Gender')
sns.boxplot(x="gender",y="writing score" ,data=data, palette="colorblind",ax=axes[2]).set_title('Writing Score by Gender')
fig.tight_layout()
plt.show()
```

### Test score breakdown by ethnicity 

```python
fig, axes = plt.subplots(1, 3)
sns.boxplot(x="race/ethnicity",y="math score" ,data=data, palette="colorblind",ax=axes[0]).set_title('Math score by ethnicity')
sns.boxplot(x="race/ethnicity",y="reading score" ,data=data, palette="colorblind",ax=axes[1]).set_title('Reading score by ethnicity')
sns.boxplot(x="race/ethnicity",y="writing score" ,data=data, palette="colorblind",ax=axes[2]).set_title('Writing score by ethnicity')
fig.tight_layout()
plt.show()
```

### Test score visualization by level of parent education

```python
fig, axes = plt.subplots(1, 3)
a = sns.boxplot(x="parental level of education",y="math score" ,data=data, palette="colorblind",ax=axes[0])
b = sns.boxplot(x="parental level of education",y="reading score" ,data=data, palette="colorblind",ax=axes[1])
c = sns.boxplot(x="parental level of education",y="writing score" ,data=data, palette="colorblind",ax=axes[2])
a.set_xticklabels(a.get_xticklabels(),rotation=330)
b.set_xticklabels(b.get_xticklabels(),rotation=330)
c.set_xticklabels(c.get_xticklabels(),rotation=330)
a.set_title('Math score by parent education')
b.set_title('Reading score by parent education')
c.set_title('Writing score by parent education')
fig.tight_layout()
plt.show()
```

### Test Score visualization by lunch aid type and preparation course

```python
fig, axes = plt.subplots(1, 3)
sns.boxplot(x="lunch",y="math score" ,data=data, hue='test preparation course', palette="colorblind",ax=axes[0]).set_title("Math score by prep course and lunch type")
sns.boxplot(x="lunch",y="reading score" ,data=data, hue='test preparation course', palette="colorblind",ax=axes[1]).set_title("Reading score by prep course and lunch type")
sns.boxplot(x="lunch",y="writing score" ,data=data, hue='test preparation course', palette="colorblind",ax=axes[2]).set_title("Writing score by prep course and lunch type")
fig.tight_layout()
plt.show()
```
